#### 4) How can you group data in a DataFrame by a specific column and apply an aggregate function? (code)

```python
# here I am showing all the other colums for nationality with count() aggregate function
grouped_data = data.groupby('Nationality').count() 
# now only age with the average
grouped_data = data.groupby('Nationality')[['Age']].mean() 
grouped_data
```

```python
#for this I am creating two dataframe for the existing dataset
data1 = data[['Name', 'Age', 'Nationality']]
data2 = data[['Club', 'Overall', 'Release Clause','Age']]

# I am doing left join for this
after_merged = pd.merge(data1, data2, how="left", on="Age")
after_merged.head()
```

#### 7)How can you remove numeric values from a text in Python? Provide an example. (code) 

```python
import re

text = "My name is, Tabib. My number is 017xxxxxxx"

cleaned_text = re.sub(r'\d+', '', text)  # \d+ matches one or more digits
cleaned_text
```

#### 8) What is the purpose of normalization in text preprocessing? Give an example. (theory + code) 

```python
import re

# Sample text
text = "I can't believe this! I love natural language processing for sentiment analysis, isn't it?"

# Normalization function
def normalize_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Expand contractions (optional, just a simple example)
    contractions = {"can't": "cannot", "it's": "it is", "isn't": "is not"}
    for contraction, full_form in contractions.items():
        text = text.replace(contraction, full_form)
    
    return text

# Normalize the text
normalized_text = normalize_text(text)
print("Normalized Text:", normalized_text)
```

# Text Vectorization:

#### 4) Describe how you can create a TF-IDF representation of a text corpus using Python. (code) 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "My name is Tabib",
    "I am a student of DIU",
    "I love natural language processing for sentiment analysis"
]

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus into a TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(corpus)

# Convert the TF-IDF matrix to an array and get feature names (terms)
tfidf_array = tfidf_matrix.toarray()
terms = vectorizer.get_feature_names_out()

# Print TF-IDF array and corresponding terms
print("TF-IDF Matrix:")
print(tfidf_array)
print("\nFeature Names (Terms):")
print(terms)
```
